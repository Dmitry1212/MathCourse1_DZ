```python

import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
%matplotlib inline
```

```python
x = np.linspace(0, 4*np.pi, 100)
k, a, b = 1, np.pi, 1
plt.plot(x, k * np.cos(x - a) + b)
k, a, b = 2, np.pi, 1
plt.plot(x, k * np.cos(x - a) + b)
k, a, b = 1, 2*np.pi, 1
plt.plot(x, k * np.cos(x - a) + b)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

```python
def pol_to_dec(r, a):
    return  r * np.cos(a), r * np.sin(a)

x, y = pol_to_dec(2, np.pi / 4)
print(x,y)
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, 1 - (1 - np.exp(x))/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,8) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def func(p):
    x, y = p
    return [y -x**2 +1,
            np.exp(x) + x * (1 - y) - 1]

x1, y1 =  fsolve(func, (-2, -2))
              
print (x1, y1)
```

```python
from scipy.optimize import fsolve
def func(p):
    x, y = p
    return [y -x**2 +1,
            np.exp(x) + x * (1 - y) - 1]
root1 = fsolve(func, (-2, -2))
root2 = fsolve(func, [3, 6])
print(root1, root2)
```
