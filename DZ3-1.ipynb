```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def add_vectors(v, w):
    return [vi + wi for vi, wi in zip(v, w)]

A, B  = [10, 10, 10], [0, 0, -10]
C = add_vectors(A, B)
print (C)
```

```python
x = []    # окружность
y1 = []
y2 = []
r = 3       # радиус
m = 1000    # масштаб
for i in range(-r*m, r*m, 1):
    x_ = i/m
    x.append(x_)
    y1.append(math.sqrt(r**2 - x_**2))
    y2.append(-math.sqrt(r**2 - x_**2))    
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-r,r,-r,r])
plt.axes().set_aspect(1)
```

```python
x = []      #эллипс
y1 = []
y2 = []
m = 1000    # масштаб
a = 6       # малая полуось a>b
b = 4       # большая полуось
for i in range(-round(a*m), round(m*a), 1):
    x_ = i/m
    x.append(x_)
    y1.append(b * math.sqrt(1 - x_**2/a**2))
    y2.append(-b * math.sqrt(1 - x_**2/a**2))    
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-a,a,-b,b])
plt.axes().set_aspect(1)
```

```python
x = []        # гипербола
y1 = []
y2 = []
r = 3       # радиус
m = 1000    # масштаб
a = 3       # малая полуось a>b
b = 2       # большая полуось
for i in range(-round(a*m), round(m*a), 1):
    x_ = i/m
    x.append(x_)
    y1.append(b * math.sqrt(x_**2/a**2 + 1))
    y2.append(-b * math.sqrt(x_**2/a**2 + 1))    
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-a*2,a*2,-b*2,b*2])
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)
Z = 3 * X + 4 * Y + 150
Z2 = 3 * X + 4 * Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```

```python
fig = figure() 
ax = Axes3D(fig)
X = np.arange(-5, 5, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(X, Y)
Z =  np.sqrt(X**2 + Y**2)         # конус 2-го порядка
Z2 = - np.sqrt(X**2 + Y**2)
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```
